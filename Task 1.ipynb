### Data Preparation

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("26dd0956-8aa9-455a-8da1-9256da952da1_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python

columns_to_drop = ['Series Name', 'Series Code']

# Remove leading/trailing whitespaces from column names
df.columns = df.columns.str.strip()

# Drop the specified columns with case-insensitive matching
df.drop(columns=[col for col in columns_to_drop if col in df.columns], axis=1, inplace=True)

df.head()
```

```python
df.info()
```

```python
# Extract the years from the original column names, starting from the third column
years = [col.split(' ')[0] for col in df.columns[2:]]

# Map the old column names (starting from the third column) to the new column names without "YR" prefix
column_mapping = {old_col: year for old_col, year in zip(df.columns[2:], years)}

# Rename the columns
df.rename(columns=column_mapping, inplace=True)

df.head()
```

```python
# Use drop to remove the row based on a condition
df = df[df['Country Name'] != 'Not classified']
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
